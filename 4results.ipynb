```python
import json
import csv
import os

def add_ids_for_predicted(inputid_file, inputpre_file, output_file):
    with open(inputid_file, 'r', encoding='utf-8') as file1:
        ids = [json.loads(line)['id'] for line in file1]
    
    with open(inputpre_file, 'r', encoding='utf-8') as file2, open(output_file, 'w', encoding='utf-8') as file2_updated:
        for i, line in enumerate(file2):
            data = json.loads(line)
            if i < len(ids): 
                data['id'] = ids[i]  
            file2_updated.write(json.dumps(data, ensure_ascii=False) + '\n')

    print(f"ID added to {output_file}")


def convert_pred_to_csv(json_file, output_csv_path):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = [json.loads(line) for line in f]

    with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)

        max_tokens_length = max(len(item["tokens"]) for item in data)
        header = ["" for _ in range(max_tokens_length)]
        writer.writerow(header)

        for item in data:
            tokens_row = ["" for _ in range(max_tokens_length)]
            for i, token in enumerate(item["tokens"]):
                tokens_row[i] = token
            writer.writerow(tokens_row)

            for span in item["predicts"]:
                row = ["" for _ in range(max_tokens_length)]
                start, end, entity_type = span[0], span[1], span[2]
                for i in range(start, end):
                    row[i] = entity_type
                writer.writerow(row)

            id_row = ["#ID"] + [item["id"]] + ["" for _ in range(max_tokens_length - 2)]
            writer.writerow(id_row)
            writer.writerow([])

    print(f"CSV saved to {output_csv_path}")

inputid_json = "/home/wp/AdaSeq-master/steps/fold/20250625data08_test.json" 
# inputpre_json = "/home/wp/AdaSeq-master/experiments0926/gp_xlmroberta_large2/251016203652.889896/pred.json" 
inputpre_json = "/home/wp/AdaSeq-master/experiments0926/gp_xlmroberta_large8/251018122729.128379/pred_test_10epoch.json"  
output_json = "test_pred0625.json"  
output_csv_path = "test_pred0625.csv"  

add_ids_for_predicted(inputid_json, inputpre_json, output_json)
convert_pred_to_csv(output_json, output_csv_path)

```

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

# Load CSV files
df = pd.read_csv("0625test_label.csv")  # True labels
df1 = pd.read_csv("0625test_pred.csv")  # Predicted labels

# Define your label pairs
label_pairs = [
    ('Act_active', 'Act_active'),
    ('Act_light', 'Act_light'),
    ('Act_stationary', 'Act_stationary'),
    
    ('F_active', 'F_active'),
    ('F_recreation', 'F_recreation'),
    ('F_stationary', 'F_stationary'),
    ('F_cultural', 'F_cultural'),
    ('F_service', 'F_service'),
    ('F_Transport', 'F_Transport'),
    ('F_catering', 'F_catering'),
    ('F_Sign', 'F_Sign'),
    ('F_path', 'F_path'),
    ('F_lights', 'F_lights'),

    
    ('N_Terrain', 'N_Terrain'),
    ('N_Plant', 'N_Plant'),
    ('N_Flower', 'N_Flower'),
    ('N_Animal', 'N_Animal'),
    ('N_Water', 'N_Water'),
    ('N_Weather', 'N_Weather'),
    ('N_Color', 'N_Color'),
    
    ('P_Sound', 'P_Sound'),
    ('P_Olf', 'P_Olf'),
    ('P_Tactile', 'P_Tactile'),
    ('P_Crowd', 'P_Crowd')
]

# Initialize a list to store the metrics for each label
metrics_list = []

# Loop over each label pair and calculate metrics
for human_label, lexicon_label in label_pairs:
    y_true = df[human_label]
    y_pred = df1[lexicon_label]
    
    # Calculate each metric
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    
    # Append the metrics to the list as a dictionary
    metrics_list.append({
        'Type': human_label,
        'LCNER_C_P': precision,
        'LCNER_C_R': recall,
        # 'LCNER_C_Acc': accuracy,
        'LCNER_C_F1': f1
    })

# Convert the list of metrics into a DataFrame
metrics_df = pd.DataFrame(metrics_list)

# Save the metrics DataFrame to a CSV file
metrics_df.to_csv("metrics_LCNER_C.csv", index=False)
metrics_df
```

```python
import os
import pandas as pd

df_merged = pd.read_csv("0625test_lexicon.csv")
df_feature_word = pd.read_csv('/home/wp/secondwork/Feature1023.csv')
df_feature_word = df_feature_word.loc[:, ~df_feature_word.columns.str.contains('^Unnamed')]
target_words = df_feature_word.dropna(how='all', axis=1).columns.tolist()

output_dir = 'lexicon_aspect'
os.makedirs(output_dir, exist_ok=True)

for col in target_words:
    output_file_path = os.path.join(output_dir, f'{col}.csv')

    target_words_list = (
        df_feature_word[col]
        .dropna()
        .astype(str)
        .str.lower()
        .unique()
        .tolist()
    )

    if not target_words_list:
        print(f"No words in lexicon for column {col}. Skipping...")
        continue

    df_merged['text_lower'] = df_merged['text'].str.lower()
    regex_pattern = r'\b(' + '|'.join(target_words_list) + r')\b'


    target_rows = df_merged.loc[
        df_merged['text_lower'].str.contains(regex_pattern, regex=True, na=False)
    ].copy()

    if target_rows.empty:
        print(f"No matches found for {col}. Skipping...")
        continue

    target_rows['feature'] = target_rows['text_lower'].apply(
        lambda x: next((word for word in target_words_list if word in x.split()), '')
    )

    target_rows = target_rows.drop(columns=['text_lower'])

    target_rows.to_csv(output_file_path, index=False)
    print(f"File '{col}.csv' has been created.")
```

```python
import os
import pandas as pd

source_dir = 'lexicon_aspect'

target_csv_path = '0625test_lexicon.csv'

output_csv_path = '0625test_lexicon1.csv'

df_target = pd.read_csv(target_csv_path, dtype={'id': str})

source_csv_files = [f for f in os.listdir(source_dir) if f.endswith('.csv')]
features_labels_df = pd.DataFrame()

for csv_file in source_csv_files:
    file_path = os.path.join(source_dir, csv_file)
    df_source = pd.read_csv(file_path, dtype={'id': str})
    
    name = os.path.splitext(csv_file)[0]
    label_col = f'Lexicon_{name}'
    df_source[label_col] = 1  

    if features_labels_df.empty:
        features_labels_df = df_source.copy()
    else:
        features_labels_df = pd.merge(
            features_labels_df,
            df_source,
            on='id',
            how='outer',
            suffixes=('', '_dup')
        )
        for col in list(features_labels_df.columns):
            if col.endswith('_dup'):
                base = col[:-4]
                features_labels_df[base] = features_labels_df[[base, col]] \
                    .apply(lambda row: ', '.join(map(str, filter(pd.notna, row))), axis=1)
                features_labels_df.drop(columns=[col], inplace=True)

features_labels_df['id'] = features_labels_df['id'].astype(str)

df_merged = pd.merge(df_target, features_labels_df, on='id', how='left')

if 'text_y' in df_merged.columns:
    df_merged.drop(columns=['text_y'], inplace=True)
if 'text_x' in df_merged.columns:
    df_merged.rename(columns={'text_x': 'text'}, inplace=True)

df_merged.fillna(0, inplace=True)

df_merged.to_csv(output_csv_path, index=False)

for col in df_merged.columns:
    if pd.api.types.is_numeric_dtype(df_merged[col]):
        print(f"{col}: {(df_merged[col] == 1).sum()}")

```

```python
import pandas as pd

df_lexicon = pd.read_csv('0625label_lexicon_output_metrics.csv')
df_llm     = pd.read_csv('0625label_LLM_output_metrics.csv')
df_lcner   = pd.read_csv('metrics_LCNER.csv')
df_lcner_c = pd.read_csv('metrics_LCNER_C.csv')

df_merged = df_lexicon.merge(df_llm, on='Type', how='outer', suffixes=('_lexicon', '_llm')) \
                     .merge(df_lcner, on='Type', how='outer', suffixes=('', '_lcner')) \
                     .merge(df_lcner_c, on='Type', how='outer', suffixes=('', '_lcner_c'))

for col in df_merged.select_dtypes(include='float'):
    df_merged[col] = df_merged[col].round(3)

def sort_cols(cols):
    main = ['Type']
    suffixes = ['_F1', '_P', '_R']
    newcols = []
    for suf in suffixes:
        matched = [c for c in cols if c.endswith(suf) and c not in main]
        newcols.extend(matched)
    rest = [c for c in cols if c not in main + newcols]
    return main + newcols + rest

df_merged = df_merged[sort_cols(df_merged.columns)]
mapping = {
        'Act_active':   'Act_MVPA',
        'Act_stationary':   'Act_sedentary',
            'F_active':   'F_MVPA',
        'F_stationary':   'F_sedentary',
    
    'F_Sign':   'F_sign',
    'F_Transport': 'F_transport',
    'N_Animal': 'N_animal',
    'N_Color':  'N_color',
    'N_Flower': 'N_flower',
    'N_Plant':  'N_plant',
    'N_Terrain':'N_terrain',
    'N_Water':  'N_water',
    'N_Weather':'N_weather',
    'P_Crowd':  'P_crowd',
    'P_Olf':    'P_olf',
    'P_Sound':  'P_sound',
    'P_Tactile':'P_tactile'
}

df_merged['Type'] = df_merged['Type'].replace(mapping)

df_merged.to_csv('merged_metrics_all.csv', index=False)
df_merged
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgba
import pandas as pd

df = pd.read_csv('merged_metrics_all.csv')
type_labels = df['Type'].tolist()

model_labels = ['Lexicon', 'Lexicon+LLM', 'LCNER-span', 'LCNER']
f1_cols = ['Lexicon_F1', 'L+LLM_F1', 'LCNER_F1', 'LCNER_C_F1']
prec_cols = ['Lexicon_P', 'L+LLM_P', 'LCNER_P', 'LCNER_C_P']
recall_cols = ['Lexicon_R', 'L+LLM_R', 'LCNER_R', 'LCNER_C_R']

viridis_colors = plt.cm.viridis(np.linspace(0.2, 0.95, 4))

def darken(color, amount=0.8):
    rgb = np.array(color[:3])
    return tuple(rgb * amount)

soft_colors = [to_rgba(c, 0.8) for c in viridis_colors] 
dark_colors = [darken(c) for c in viridis_colors]          


width = 0.18
x = np.arange(len(type_labels))
fig, axes = plt.subplots(3, 1, figsize=(16, 13), sharex=True)

metric_cols = [f1_cols, prec_cols, recall_cols]
metric_names = ["F1 Score", "Precision", "Recall"]

for k, ax in enumerate(axes):
    cols = metric_cols[k]

    for i, col in enumerate(cols):
        ax.bar(x + i*width, df[col], width, color=soft_colors[i])

    for j in range(len(type_labels)):
        val0 = df[cols[0]][j]
        val3 = df[cols[3]][j]
        delta = val3 - val0
        ax.text(x[j], val0 - 0.09, f"{val0:.3f}", ha='center', va='top', fontsize=11, color=dark_colors[0], fontweight='bold')
        ax.text(x[j] + 3*width, val3 + 0.04, f"{delta:+.3f}", ha='center', va='bottom', fontsize=11, color=dark_colors[3], fontweight='bold')

    ax.set_ylabel(metric_names[k], fontsize=14)
    ax.grid(True, axis='y', linestyle='--', alpha=0.4)

    ymax = max(df[cols[0]].max(), df[cols[3]].max())
    ax.set_ylim(0, ymax + 0.1)
    ax.set_xlim(-0.5, len(type_labels) - 0.3 + 2 * width)


axes[-1].set_xticks(x + 1.5*width)
axes[-1].set_xticklabels(type_labels, rotation=30, ha='right', fontsize=14)

handles = [plt.Rectangle((0, 0), 1, 1, color=soft_colors[i]) for i in range(4)]
fig.legend(
    handles, model_labels,
    loc='upper center',
    ncol=4,
    fontsize=14,
    bbox_to_anchor=(0.5, 0.95),
    frameon=False
)

plt.tight_layout(rect=[0, 0.04, 1, 0.93])
plt.subplots_adjust(hspace=0.08)
plt.savefig("4Results.png", dpi=300, bbox_inches='tight')
plt.show()
```
